## ตารางข้อมูล covid ปี2022

```python
data_covid2022 =  pd.read_csv(os.path.join(path,'confirmed-cases-since-280265.csv'))
data_covid2022
```

## ต่อตาราง


```python
covid = pd.concat([data_covid2020,data_covid2021_1,data_covid2021_2,data_covid2022])
covid
```

## เลือกเฉพาะคอลัมน์ที่สนใจ

```python
covid_data = covid[['No.','announce_date','sex','age','Unit','nationality','risk','province_of_onset']]
covid_data
```

## ตรวจสอบค่า Missing

```python
covid_data.isnull().any() #ดูว่าแต่ละ column มีค่า missing มั้ย
```

```python
covid_data[['sex']] = covid_data[['sex']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid_data[['Unit']] = covid_data[['Unit']].fillna('ปี') ##แทนค่า NaN ใน Unit เป็น ปี
covid_data[['nationality']] = covid_data[['nationality']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid_data[['province_of_onset']] = covid_data[['province_of_onset']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid_data[['risk']] = covid_data[['risk']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
```

```python
covid_data = covid_data.fillna(method="bfill") #ใช้ bfill คือการแทนที่ด้วยค่าถัดไปคือการนำค่าถัดไปที่มีค่า ของเซลล์ตัวเองมาแทนที่ลงในช่องที่เป็น NaN
covid_data.isnull().any() #เช็คข้อมูลในแต่ละคอลัมน์อีกครั้ง
```
